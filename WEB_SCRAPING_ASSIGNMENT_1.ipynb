# Q.9 Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
restruant_name = []
```

```python
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    restruant_name.append(i.text)
restruant_name
```

```python
restruant_location = []
```

```python
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    restruant_location.append(i.text)
restruant_location
```

```python
restruant_images = []
```

```python
for i in soup.find_all('img',class_="no-img"):
    restruant_images.append(i.get('data-src'))
restruant_images
```

```python
rest_rating = []
```

```python
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rest_rating.append(i.text)
rest_rating
```

```python
rest_cuisine = []
```

```python
for i in soup.find_all('span',class_="double-line-ellipsis"):
    rest_cuisine.append(i.text.split('|')[1])
rest_cuisine
```

```python
import pandas as pd
df=pd.DataFrame({'Restaurant Name':restruant_name,'Cuisines':rest_cuisine,'Location':restruant_location,'Ratings':rest_rating,'Image URL':restruant_images})
df
```

# Q.8 Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
title = []
```

```python
for i in soup.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    title.append(i.text)
title
```

```python
authors = []
```

```python
for i in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    authors.append(i.text)
authors
```

```python
published_date=[]
```

```python
for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    published_date.append(i.text)
published_date
```

```python
paper_URL = []
```

```python
for i in soup.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    paper_URL.append(i.get('href'))
paper_URL
```

```python
import pandas as pd
df = pd.DataFrame({'Paper Title':title,'Authors':authors,'Published Date':published_date,'Paper URL':paper_URL})
df
```
